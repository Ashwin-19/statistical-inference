$\text{First, we obtain our sample from the one given in the assignment.}$

```r
sample <- c(10.4, 19.6, 18.8, 13.9, 17.8, 16.8, 21.6, 17.9,12.5, 11.1, 4.9, 12.8, 14.8, 22.8, 20.0,
15.9,16.3, 13.4,17.1,14.5, 19.0, 22.8, 1.3, 0.7, 8.9, 11.9, 10.9, 7.3, 5.9, 3.7, 17.9, 19.2,
9.8, 5.8, 6.9, 2.6, 5.8, 21.7, 11.8, 3.4, 2.1, 4.5, 6.3, 10.7, 8.9, 9.4, 9.4, 7.6, 10.0, 3.3,
6.7, 7.8, 11.6, 13.8, 18.6)
```

$\text{(a) Since the observations follow uniform}(0,\theta)\text{ distribution, the likelihood function is}$
$L(\theta|x_{1},.,..x_{n}) =  \begin{cases} 
      (\frac{1}{\theta})^{n} & 0 \leq x_{i} \leq \theta\\
      0 & otherwise \\
   \end{cases}
$
<br/>
$\text{And the log-likelihood function can be written as}$
<br/>
$l(\theta|x_{1},.,..x_{n}) = -n\log(\theta)\text{ for }0 \leq x_{i} \leq \theta$
$\text{We maximize this function to yield estimates for }\theta.$

```r
# log likelihood function
loglikelihood <- function(theta){
    
    # no. of values in the sample
    n <- length(sample)
    
    # compute likelihood using the formula
    likelihood <- -n*log(theta)
    
    return(likelihood)
}
```

```r
optim(par=min(sample),fn=loglikelihood,gr=NULL,method='L-BFGS-B',lower=min(sample),upper=max(sample))
```

$\text{Therefore, we obtain maximum likelihood estimate of }\theta \text{ to be }22.8$

$\text{(b) We plot the likelihood function for different values of }\theta$

```r
likelihood <- function(theta){
    
    # no. of values in the sample
    n <- length(sample)
    
    # check for values not in range
    for(i in sample){
        if(i > theta){
            return(0)
        }
    }
    
    # compute likelihood using the formula
    likelihood <- (1/theta)^n
    
    return(likelihood)
}

theta_vec <- seq(min(sample),max(sample)+10,by=0.1)
likelihood_vec <- seq(min(sample),max(sample)+10,by=0.1)

i <- 1
for(theta in theta_vec){
    likelihood_vec[i] <- likelihood(theta)
    i <- i + 1
}

plot(theta_vec, likelihood_vec,xlab='theta',ylab='likelihood',main='likelihood function vs theta',col='blue')
abline(v=22.8,col='black')
text(22.8,0,"22.8",col='black',adj = c(-.1, -.1))
```

$\text{As we can see, the function attains maximum value at theta = 22.8}$
