$\text{First, we generate a random sample of size 1000 from Geometric (p = 0.6).}$

```r
sample <- rgeom(1000, prob = 0.6)
```

$\text{(a) Since the sample follows Geometric(p = 0.6), the likelihood function can be written as}$
$L(\theta|x_{1},.,..x_{n}) = \prod_{i=1}^{n} f_{i}(x_{i}|\theta) = \prod_{i=1}^{n}(1-p)^{x_{i}}p = (1-p)^{\sum_{i=1}^{n} x_{i}}p^{n}$
$\text{Now, the log-likelihood function can be written as}$
<br/>
$l(\theta|x_{1},.,..x_{n}) = \log((1-p)^{\sum_{i=1}^{n} x_{i}}p^{n})$
$\implies l(\theta|x_{1},.,..x_{n}) = (\sum_{i=1}^{n} x_{i})\log{(1-p)} + n\log{p}$
$\text{We create a negative version of this log-likelihood function and yield estimates upon}$ 
$\text{minimizing it.}$

```r
# calculate -loglikelihood for geometric(p)
loglikelihood <- function(p){
    
    # compute length of sample
    n <- length(sample)
    
    # calculate sum of xi's
    sum <- sum(sample)
    
    # compute log likelihood using the formula
    likelihood <- sum*log(1-p) + n*log(p)
    
    # return -ve of likelihood
    return(-likelihood)
}

MLE = suppressWarnings(nlminb(start=0.1,objective=loglikelihood,hessian=T))
MLE
```

$\text{As we can see, the maximum likelihood estimate of p is 0.59 and the corresponding}$
$\text{value of the log likelihood function is 1145.50}$

$\text{(b) To verify the results, we plot a log-likelihood graph for various values of p.}$

```r
# calculate loglikelihood for geometric(p)
loglikelihood <- function(p){
    
    # compute length of sample
    n <- length(sample)
    
    # calculate sum of xi's
    sum <- sum(sample)
    
    # compute log likelihood using the formula
    likelihood <- sum*log(1-p) + n*log(p)
    
    # return -ve of likelihood
    return(likelihood)
}

# initialize values of p
p_vec <- seq(0.1,0.9,by=0.005)
loglikelihood_vec <- seq(0.1,0.9,by=0.005)

# calculate log-likelihood for each value
for(i in 1:length(p_vec)){
    loglikelihood_vec[i] <- loglikelihood(p_vec[i])
}
```

```r
# plot log-likelihood for various values of p
plot(p_vec,loglikelihood_vec,xlab='p',ylab='log-likelihood',main='log-likelihood vs p')
text(0.59,0,"p = 0.59",col=2)
abline(v=0.59,col='blue')
```

$\text{As we can see, the function attains maximum value at p = 0.59}$
